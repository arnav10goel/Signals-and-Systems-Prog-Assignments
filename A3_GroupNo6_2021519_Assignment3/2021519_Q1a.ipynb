```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

### Plotting x[n]:

```python
def x1(n):
    if n == 0:
        return 1
    else:
        return 0

a = np.array([n for n in range(-1000, 1001)])
b = np.array([x1(n) for n in range(-1000, 1001)])

plt.stem(a, b)
plt.xlabel('n')
plt.ylabel('x1[n]')
plt.show()
```

### Real Part of Complex Discrete Time Fourier Transform

```python
def dft_real(w):
    res = 0
    for i in a:
        res += x1(i)*math.cos(-w*i)
    return res

arr_w = np.linspace(-2*math.pi,2*math.pi,1000)
arr_dtf_real = np.array([dft_real(w) for w in arr_w])

plt.plot(arr_w, arr_dtf_real)
plt.xlabel('w')
plt.ylabel('real{X(e^jw)}')
plt.show()
```

### Imaginary Part of Complex Discrete Time Fourier Transform

```python
def dft_img(w):
    res = 0
    for i in a:
        res += x1(i)*math.sin(-w*i)
    return res

arr_dtf_img = np.array([dft_img(w) for w in arr_w])

plt.plot(arr_w, arr_dtf_img)
plt.xlabel('w')
plt.ylabel('imag{X(e^jw)}')
plt.show()
```

### Magnitude Spectrum of DFT Signal

```python
def dft_mag(w):
    a = dft_real(w)
    b = dft_img(w)
    return math.sqrt(a**2 + b**2)

arr_dtf_mag = np.array([dft_mag(w) for w in arr_w])

plt.plot(arr_w, arr_dtf_mag)
plt.xlabel('w')
plt.ylabel('mag{X(e^jw)}')
plt.show()
```
