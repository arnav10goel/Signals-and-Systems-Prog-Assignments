```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

### Plotting x[n]:

```python
def x2(n):
    if n <= 4 and n >= -4:
        return 1
    else:
        return 0

a = np.array([n for n in range(-1000, 1001)])
b = np.array([x2(n) for n in range(-10, 11)])

c = np.array([n for n in range(-10, 11)])

plt.stem(c, b)
plt.xlabel('n')
plt.ylabel('x2[n]')
plt.show()
```

### Real Part of Complex Discrete Time Fourier Transform

```python
def dft_real(w):
    res = 0
    for i in c:
        res += x2(i)*math.cos(-w*i)
    return res

arr_w_1 = np.linspace(-2*math.pi,2*math.pi,1000)
arr_dtf_real = np.array([dft_real(w) for w in arr_w_1])

plt.plot(arr_w_1, arr_dtf_real)
plt.xlabel('w')
plt.ylabel('real{X(e^jw)}')
plt.show()
```

### Imaginary Part of Complex Discrete Time Fourier Transform

```python
def dft_img(w):
    res2 = 0
    for i in c:
        res2 += (x2(i))*(math.sin(-w*i))
    return res2

arr_w_2 = np.linspace(-2*math.pi,2*math.pi,1000)
arr_dtf_img = np.array([dft_img(w) for w in arr_w_2])

plt.plot(arr_w_2, arr_dtf_img)
plt.xlabel('w')
plt.ylabel('imag{X(e^jw)}')
plt.ylim(-1,1)
plt.show()
```

### Magnitude Spectrum of DFT Signal

```python
def dft_mag(w):
    a = dft_real(w)
    b = dft_img(w)
    return math.sqrt(a**2 + b**2)

arr_dtf_mag = np.array([dft_mag(w) for w in arr_w_1])

plt.plot(arr_w_1, arr_dtf_mag)
plt.xlabel('w')
plt.ylabel('mag{X(e^jw)}')
plt.show()
```
