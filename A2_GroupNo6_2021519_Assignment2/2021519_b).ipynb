Plotting h(t):

```python
import math
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate

def u(n):
    if n < 0:
        return 0
    else:
        return 1

def h(t):
    y = u(t)
    x = (0.50)*((math.e)**((-1)*(t))-(math.e)**((-4)*(t)))
    return y*x

arr = np.linspace(0,20,1000)
res1 = []
for i in arr:
    res1.append(h(i))
fig = plt.figure()
plt.plot(arr, res1)
plt.title("Plotting h(t)")
plt.xlabel('t')
plt.ylabel('h(t)')
plt.show()
```

Plotting x(t):

```python
def x(t):
    return (math.sin(t))*u(t)*((math.e)**(-t))
res2 = []
for i in arr:
    res2.append(x(i))
fig = plt.figure()
plt.plot(arr, res2)
plt.title("Plotting x(t)")
plt.xlabel('t')
plt.ylabel('x(t)')
plt.show()
```

Plotting y(t) - Convolving x(t) and h(t):

```python
res_h = np.array(res1)
res_x = np.array(res2)

def integrand(t0, t):
    return x(t0)*h(t-t0)

res_y = []
for t in arr:
    res_y.append(integrate.quad(integrand, 0, 20, args=t)[0])

fig = plt.figure()
plt.plot(arr, res_y)
plt.title("Convolution of x(t) and h(t)")
plt.xlabel('t')
plt.ylabel('y(t)')
plt.show()
```
